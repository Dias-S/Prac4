```python
import json

file_path = "sample-data.json"


with open(file_path, "r", encoding="utf-8") as file:
    data = json.load(file)


print(json.dumps(data, indent=4))
```

```python
if "imdata" in data:  
    interfaces = data["imdata"]  
else:
    print("Error: In JSON doesn't 'imdata'")
    interfaces = []


print(f"{'DN':<50} {'Description':<20} {'Speed':<10} {'MTU':<10}")
print("=" * 90)


for item in interfaces:
    interface_info = item.get("l1PhysIf", {}).get("attributes", {})

    dn = interface_info.get("dn", "N/A")
    description = interface_info.get("descr", "N/A")
    speed = interface_info.get("speed", "N/A")
    mtu = interface_info.get("mtu", "N/A")

    print(f"{dn:<50} {description:<20} {speed:<10} {mtu:<10}")
```

```python
import json


file_path = "sample-data.json"


with open(file_path, "r", encoding="utf-8") as file:
    data = json.load(file)


interfaces = data.get("imdata", [])


print("Interface Status")
print("=" * 90)
print(f"{'DN':<50} {'Description':<20} {'Speed':<10} {'MTU':<10}")
print("-" * 90)


for item in interfaces:
    interface_info = item.get("l1PhysIf", {}).get("attributes", {})

    dn = interface_info.get("dn", "N/A")
    description = interface_info.get("descr", "N/A")
    speed = interface_info.get("speed", "N/A")
    mtu = interface_info.get("mtu", "N/A")

    print(f"{dn:<50} {description:<20} {speed:<10} {mtu:<10}")
```

```python
import csv

with open("output.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["DN", "Description", "Speed", "MTU"])
    
    for item in interfaces:
        interface_info = item.get("l1PhysIf", {}).get("attributes", {})
        writer.writerow([
            interface_info.get("dn", "N/A"),
            interface_info.get("descr", "N/A"),
            interface_info.get("speed", "N/A"),
            interface_info.get("mtu", "N/A")
        ])
    
print("Saved in.csv")
```
